```python
## Measuring Success: Training and Testing data
```

```python
# Let's call the train_test_split on our data and assign the outputs using this nomenclature
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_datasets['data'], iris_datasets['target'], random_state = 0)
```

```python
print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
```

```python
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

```python
## First Things First: Look at your data(By Data Visulization)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_datasets.feature_names)
```

```python
pd.plotting.scatter_matrix(iris_dataframe, c = y_train, figsize = (12,15),
                          marker = 'o', hist_kwds = {'bins':20},
                          s = 60, alpha = .8, cmap = mglearn.cm3)
```

```python
## Buliding Model: K-Nearest Neighbors
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
## Making Prediction
```

```python
x_new = np.array([[5,2.9,1,0.2]])
print("X_new.shape:  {}".format(x_new.shape))
```

```python
prediction = knn.predict(x_new)
print('Predition: {}'.format(prediction))
print('Prediciton target name: {}'.format(iris_datasets['target_names'][prediction]))
```

```python
## Elevating the model
```

```python
y_pred = knn.predict(X_test)
print('Test set prediction: \n {}'.format(y_pred))
```

```python
print('Test set score: {:.2f}'.format(np.mean(y_pred == y_test)))
```

```python
print('Test set score: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
## Summary and Outlook
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_datasets['data'], iris_datasets['target'], random_state = 0)
```
